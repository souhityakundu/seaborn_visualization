```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
data = sns.load_dataset("tips")
data.head()
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data = data)
```

```python
# LETS BRING ON STYLE
sns.relplot(x="total_bill",y="tip",hue="smoker",style="sex",data = data)
```

```python
# style,hue all are categorical data
sns.relplot(x="total_bill",y="tip",hue="smoker",size="size",style="sex",data = data)
```

#CATEGORICAL DATA

```python
# catplot:
# striplots
# swarmplots
# x-->categorical datasets
# y-->numerical   datasets
plt.style.use("fivethirtyeight")
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="strip")
```

```python
sns.catplot(x="day",y="tip",data=tips,kind="swarm")
```

# BOXPLOT

```python
#UNIVARIATE
sns.boxplot(tips["total_bill"])
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="box",hue="sex")
```

```python
sns.catplot(x="day",y="total_bill",hue="sex",kind="violin",split=True,data=tips)
```

# BARPLOT AND COUNTPLOT

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="bar")
```

```python
sns.catplot(x="sex",kind="count",hue="smoker",data=tips)
```

# DISTPLOT

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic["age"].fillna(titanic["age"].mean(),inplace=True)
titanic.isnull().sum()
titanic.sample(20)
```

```python
sns.distplot(titanic["age"])
```

```python
sns.displot(titanic["age"])
```

```python
sns.kdeplot(titanic["age"])
```

```python
sns.distplot(titanic["age"],hist=False,kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic["survived"]==1]["age"])
sns.distplot(titanic[titanic["survived"]==0]["age"])
```
